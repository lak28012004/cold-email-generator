```python
%%writefile app.py

from flask import Flask, render_template, request
import random

app = Flask(__name__)

EMAIL_TEMPLATES = [
    Subject: Application Inquiry â€“ {role} at {company}

Dear {receiver},

I hope this message finds you well. My name is {sender}, and I am currently working as a {role}
with hands-on experience in {skills}.

I have been following {company} and truly admire your work culture, innovation, and impact
in the industry. I strongly believe my technical background and learning mindset would allow
me to contribute effectively to your team.

I would be grateful for the opportunity to discuss any suitable openings or internships
that align with my profile.

Thank you for your time and consideration.

Kind regards,
{sender}



]

@app.route("/", methods=["GET", "POST"])
def index():
    email = None
    if request.method == "POST":
        sender = request.form["sender"]
        receiver = request.form["receiver"]
        role = request.form["role"]
        skills = request.form["skills"]
        company = request.form["company"]

        template = random.choice(EMAIL_TEMPLATES)
        email = template.format(
            sender=sender,
            receiver=receiver,
            role=role,
            skills=skills,
            company=company
        )

    return render_template("index.html", email=email)

if __name__ == "__main__":
    app.run(port=5000)
```

```python
%%writefile run.py

from pyngrok import ngrok
import threading
from app import app

# Kill any existing tunnels (important in Colab)
ngrok.kill()

# Start ngrok tunnel
public_url = ngrok.connect(5000)
print("ðŸŒ Public URL:", public_url)

# Run Flask app
threading.Thread(
    target=app.run,
    kwargs={"host": "0.0.0.0", "port": 5000}
).start()
```
